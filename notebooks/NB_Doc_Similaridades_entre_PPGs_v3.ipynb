Etapas do processo de trabalho
==============================

Nessa seo,  apresentado o processo de trabalho que foi empregado para
prover  Diretoria de Avaliao da CAPES (DAV) indicadores no
subjetivos de casos de PPGs geograficamente prximos que produziram nos
ltimos anos teses e dissertaes em campos, temas, problemas e assuntos
com semelhana estatisticamente mensurvel. Para a determinao desses
casos foi empregado um processo de trabalho que envolveu a combinao de
mtodos e tcnicas de aprendizado de mquina e sua aplicao  grandes
volumes de dados no estruturados em forma de texto livre. Tanto as
etapas desse processo de trabalho como as tcnicas empregadas so
brevemente descritas ao longo do texto que se segue.

Processamento dos textos
------------------------

O primeiro passo consistiu em se escolher dados capazes de representar a
temtica de cada PPG para que neles se pudesse buscar os casos com maior
coincidncia. Os dados escolhidos foram as 463.164 mil teses e
dissertaes defendidas ao longo de 2013 a 2018 por todos os 4.190 PPGs
que se encontravam ativos durante esse perodo (ver Tabela 1). As teses
e dissertaes de cada PPG foram consideradas o contedo temtico e
semntico capaz de melhor representar a vocao, objetivo e foco
investigativo e cientfico de cada PPG porque so intrinsecamente
concordantes com os interesses e formao de seus docentes orientadores,
e necessariamente alinhados com as respectivas linhas de pesquisa e
reas de concentrao. Nenhum contedo textual e semanticamente rico
disponvel pode representar melhor as temticas de interesse cientfico
de docentes e programas.

Apesar da boa qualidade dos metadados de teses e dissertaes fornecidos
pela DAV, houve assim mesmo a necessidade de se realizar vrios
procedimentos de limpeza, complementao, padronizao e homogeneizao
dos dados, alm da verificao das relaes de cada tese com os PPGs,
docentes e reas corretas. As teses e dissertaes sem metadados em
ingls tiveram que ser ignoradas. A escolha do ingls como lngua dos
metadados (textos) utilizados se deveu ao fato de 1) os algoritmos e
pacotes de software para essa lngua serem muito mais avanados e
precisos do que seus primos para o Portugus, e 2) pelo fato de modelos
semnticos criados a partir de textos em ingls permitirem a incluso de
novos textos de outras fontes de qualidade (e com isso se obter um ganho
em preciso).

Utilizando pacotes de programas de **Processamento de linguagem
natural**, foram filtradas todas as palavras com significado relevante e
utilidade para a tarefa de determinao dos assuntos e temticas dos
respectivos textos. As palavras mais relevantes e representativas (ex.:
frequentes, raras, sujeitos de oraes etc.) foram ento convertidas em
representaes estatsticas e usadas para caracterizar cada tese e
dissertao, ao mesmo tempo em que se preservou a proximidade de
ocorrncia de cada termo e suas vizinhanas. O princpio que rege essa
estratgia  que palavras que ocorrem dentro de oraes prximas ou nas
mesmas oraes invariavelmente pertencem ao mesmo assunto ou tema.
Especial cuidado foi dado para no se ignorar termos compostos com
significados relevantes para suas respectivas reas de conhecimento
(ex.: Big Data para a computao).

Representao distribuda dos textos
------------------------------------

A segunda etapa consistiu na aplicao da tcnica **Representaes
distribudas de textos completos** aos textos processados na etapa
anterior. Nela foram geradas representaes espaciais distribudas para
cada palavra, resumo de tese, dissertao e respectivos docentes
orientadores e PPGs. PPGs que no receberam uma representao espacial
prpria foram somente aqueles que no estavam ativos no perodo
compreendido, no existiam antes de 2013, s passaram a existir aps
2018, ou no possuam teses e dissertaes com metadados aproveitveis,
i.e. legvel, semanticamente correto e coerente, expressivo.

Em Aprendizado de Mquina, aprender ou gerar uma representao
distribuda  um conjunto de tcnicas capaz de descobrir automaticamente
representaes que podem ser usadas para a deteco ou a classificao
dos recursos a partir de dados brutos. Isso permite que uma mquina
aprenda os recursos (usando uma representao funcional prpria) e os
use para executar uma tarefa especfica. O aprendizado de representaes
 motivado pelo fato de que as tarefas de aprendizado de mquina, como
classificao, geralmente requerem entradas que so matematicamente e
computacionalmente convenientes para serem processadas. No entanto,
dados do mundo real, como imagens, vdeo e textos, no so convenientes.
Para examin-los e us-los em algoritmos,  necessrio que se descubram
representaes coerentes para eles.

Uma representao distribuda  ento, em sntese, um conjunto de
nmeros que podem ser usados para representar o objeto (vdeo, texto,
imagem etc.) a partir do qual foi criada. No presente trabalho, as
representaes aprendidas foram usadas como coordenadas em um espao
multidimensional. Essa tcnica de aprendizado capaz de produzir
representaes distribudas de palavras foi desenvolvida para se
determinar automaticamente sinnimos e antnimos, e recebeu o nome
**Representao distribuda de palavras.**

A tcnica **Representaes distribudas de textos completos,**
efetivamente usada no presente trabalho,  uma derivao desta tcnica e
guarda as mesmas propriedades que a original, como explicado por (LE;
MIKOLOV, 2014). Ela permite se determinar, em uma escala de 0 a 1, uma
medida da semelhana relativa entre dois agregados de palavras, ou
melhor, entre dois textos de qualquer tamanho. Outra caracterstica
importante dessa tcnica  que com ela se pode gerar representaes
espaciais equivalentes s geradas para as palavras individuais dos
mesmos textos. Qualquer palavra ou agregado de palavras que compartilha
o mesmo espao semntico do modelo que criou as representaes pode ser
comparada e sua semelhana aferida. No presente trabalho, as
representaes foram geradas para os agregados de palavras resultantes
da unio dos textos dos ttulos, resumos e palavras-chaves de cada tese
e dissertao, da unio das teses e dissertaes orientadas por cada
docente e de cada PPG.

Cabe ressaltar que cada representao de uma tese, dissertao, docente
ou PPG  totalmente independentemente e no influenciada por qualquer
taxonomia ou outro tipo de representao formal atribuda pelo homem e
externa ao modelo e aos textos processados, como rea de avaliao, ou
UF, ou rea de conhecimento. As representaes mencionadas aqui, e
usadas no processo de trabalho, derivaram nica e exclusivamente dos
textos (metadados dos ttulos, resumos e palavras-chaves em ingls) das
teses, dissertaes processadas.

O principal recurso criado e utilizado nessa etapa foi um modelo
probabilstico (mais especificamente, a rede neural) usado para geradas
as representaes distribudas correspondente a cada PPG. Essas
representaes foram, por sua vez, interpretadas como coordenadas num
espao abstrato criado pelo modelo. Coordenadas e representaes
distribudas tm aqui o mesmo significado. A esse espao abstrato criado
pela reunio de todas as representaes chamamos aqui de espao
semntico, pois, em ltima instncia, representa a semntica processada
de todos os textos aproveitados.

Clculo da similaridade entre PPGs
----------------------------------

Como a representao de cada PPG ocupa uma posio dentro do espao
semntico, usou-se uma das vrias mtricas de distncia (Euclidiana)
para medir o afastamento entre essas posies. Essa distncia foi
escolhida como o indicador de semelhana, ou melhor, de similaridade
semntica entre duas entidades. A unidade dessa distncia no 
comparvel a uma unidade convencional de espao fsico, como
centmetros, milmetros ou polegadas. Trata-se de uma medida vlida
somente dentro desse espao multidimensional hipottico criado pelo
modelo. Ela varia entre 0 e 1, ou entre 0% e 100% de similaridade.
Ilustrando, dois textos idnticos, que por isso tm coordenadas
idnticas de suas respectivas localizaes dentro do espao semntico,
tero a medida de similaridade igual a 1 (ou 100% coincidentes),
enquanto dois pontos totalmente opostos, ou seja, de dois textos de
significado absolutamente diferentes (se  que isso existe) tero uma
distncia semntica de 0, ou seja, sero 0% semelhantes.

Foi ento calculada a distncia entre as representaes (ou coordenadas)
de todos os PPG. Essa medida que, como j dito, corresponde 
similaridade entre as representaes, foi usada no trabalho como o
principal indicador da probabilidade de semelhana temtica entre cada
dupla de PPGs. Como resultado prtico para o trabalho, os PPGs com
coordenadas (indicadas pelas representaes) relativamente prximas
foram considerados semanticamente mais similares e os PPGs com
coordenadas relativamente distantes menos similares. Todas as distncias
foram normalizadas para um valor entre 0 e 1. A medida de similaridade
(semntica) entre dois PPGs foi ento obtida a partir do mdulo de 1
menos a distncia entre as coordenadas dos dois PPGs, como exibido em
(a):

**Similaridade = \| 1  Distncia entre representaes \|** (a)

A validade do modelo criado e das representaes geradas, primeiramente,
foi verificada empiricamente a partir da leitura de dezenas de resumos
de teses. A coerncia entre textos com similaridade alta e sua relao
com as reas de conhecimento dos respectivos PPGs tambm foi verificada.
Em seguida, a coerncia entre as medidas de similaridade e as efetivas
semelhanas temticas entre teses e PPGs foi submetida e atestada por
docentes especialistas de diferentes reas de conhecimento.

Outra verificao realizada foi a contagem de termos tcnicos (muitos
aproveitados das palavras-chaves das prprias teses) compartilhados
pelas teses com alta medida de similaridade entre si. Observou-se que as
teses de mesmos orientadores e mesmas linhas de pesquisa se localizavam
prximas (ou com alta similaridade) e compartilhavam um nmero
significativo de termos e sinnimos. Por fim, o modelo probabilstico
criado foi submetido a um teste padro com vocabulrio genrico
(Wordnet[1]) e obteve mdia de acerto superior ao baseline do estado da
arte nesse campo.

Percentis de similaridades
--------------------------

A faixa de similaridades pode variar muito em razo da quantidade de
texto usada para representar cada entidade, no caso, PPGs, teses,
dissertaes, reas etc. Quanto menos a quantidade de texto envolvida,
maior a preciso das distncias entre duas entidades, e mais prximas ou
distantes elas podero estar. Exemplificando, para um PPG ser muito
semelhante a outro, seria necessrio que os seus contedos textuais
fossem praticamente idnticos, o que no existe. As similaridades entre
palavras j pode ser muito alta, quanto maior for a semelhana de
significado e de emprego de ambas.

Visando facilitar a identificao dos casos de similaridade elevada ou
baixa, especificamente entre PPGs e respectivas produes, optou-se por
adotar percentis da faixa de similaridades encontradas entre essas
entidades. Similaridades mais baixas, entre as encontradas, correspondem
aos percentis mais baixos, enquanto as mais altas, aos percentis
superiores.

Aprendizado de Mquina (ML)
---------------------------

Tambm chamado aprendizado automtico, essa famlia de mtodos forma um
subcampo da Engenharia e da Cincia da Computao que evoluiu a partir
dos estudos de reconhecimento de padres e da teoria do aprendizado
computacional em inteligncia artificial[3]. Os algoritmos que pertencem
a esse subcampo se caracterizam por construir modelos a partir de
entradas amostrais de dados que tm como propsito fazer previses ou
apoiar decises que sejam amparadas ou guiadas exclusivamente pelos
dados fornecidos. Ou seja, sem nenhuma subjetividade introduzida por
indivduos. Essas tcnicas se opem aos mtodos que, por exemplo, seguem
instrues e regras programadas por humanos.

O aprendizado de mquina tem como uma de suas caractersticas marcantes
a extrao de regras e padres a partir de grandes conjuntos de dados.
Quanto maior o conjunto e sua qualidade, maior tende a ser a preciso
dos modelos produzidos. O aprendizado de mquina  usado em uma
variedade de tarefas computacionais onde criar e programar algoritmos
explcitos  impraticvel. Exemplos de aplicaes incluem filtragem de
spam, reconhecimento tico de caracteres (OCR), processamento de
linguagem natural, motores de busca, diagnsticos mdicos,
bioinformtica, reconhecimento de fala, reconhecimento de escrita, viso
computacional e locomoo de robs.

No campo da anlise de dados, o aprendizado de mquina  um mtodo usado
para planejar modelos complexos e algoritmos destinados  anlise
preditiva, ou seja, para permitir que pesquisadores, cientistas de
dados, engenheiros, e analistas possam, a partir da descoberta de
relaes, padres e tendncias no evidentes (usualmente pelo volume de
dados envolvidos), "produzir decises e resultados confiveis, e que
possam ser repetidos" (FRIEDMAN, 1998).

Processamento de Linguagem Natural (PLN)
----------------------------------------

Processamento de lngua natural (PLN)  uma subrea da cincia da
computao, inteligncia artificial e da lingustica que estuda os
problemas da gerao e compreenso automtica de lnguas humanas
naturais (WIKIPEDIA.ORG, 2020).

Minerao de Texto
------------------

A minerao de texto foi outra tcnica empregada nesse trabalho. Seu
emprego teve o intuito de, ao converter texto livre em representaes
aproveitveis pelas tcnicas de aprendizado de mquina, descrito na
seo anterior. O objetivo maior da minerao de texto foi transformar o
texto extrado das teses e dissertaes em dados prprios s anlises
preditivas dos mtodos de aprendizado de mquina. Essa transformao
recorreu  aplicao de recursos, i.e., programas e algoritmos de
processamento de linguagem natural (PLN).

A minerao de texto, conhecida tambm como minerao de dados textuais,
busca extrair informaes teis de um texto. teis, no presente caso,
com o sentido de aproveitveis por outros mtodos, como padres ou
tendncias. Geralmente, a minerao de texto, com o apoio do PLN,
procede com a identificao das estruturas gramatical e semntica dos
textos, o que inclui o reconhecimento de caractersticas lingusticas
inerentes  lngua dos textos usados. A anlise dos textos,
invariavelmente, recorre  anlise lexical, o que permite o
reconhecimento de palavras e, por conseguinte, o estudo, por exemplo, da
frequncia de distribuio de palavras num texto, o reconhecimento de
padres de escrita (arranjo de palavras), identificao/anotao.

Entre as tarefas em que a minerao de texto  mais usada esto
categorizao e agrupamento de textos, reconhecimento de
conceitos/entidades (ex.: nome de personalidades, cidades, produtos),
produo de taxonomias, produo de resumos e modelagem de relaes
entre entidades. Uma aplicao comum  examinar um conjunto de
documentos escritos em uma determinada lngua e, ou modelar o conjunto
de documentos para fins de classificao preditiva ou preencher um banco
de dados ou ndice de pesquisa com as informaes extradas
(WIKIPEDIA.ORG, 2019).

Representao de distribuies de palavras (RDP)
------------------------------------------------

Recentemente, houve muitos avanos no desenvolvimento de tcnicas para a
gerao modelos preditivos para uso com linguagem. Uma dessas tcnicas,
empregada nesse trabalho, consiste em se gerar representaes
distribudas para as palavras de um texto na forma de vetores de valores
contnuos que, de forma muito simplificada, podem ser interpretadas como
localizaes em um espao semntico. Trata-se de modelos de linguagem
baseados em rede neural (MLRN). Um dos que se tornou popular foi o
proposto por (BENGIO et al., 2003). Nele, uma rede neural formada por
uma camada de projeo linear e uma camada oculta no-linear so
utilizadas para gerar a representao distribuda das palavras.

Uma outra arquitetura importante de MLRN, e que foi especificamente a
utilizada neste trabalho, foi a apresentada por (MIKOLOV et al., 2013),
pela qual primeiro se aprende os vetores representando palavras
utilizando por meio de rede neural com uma nica camada oculta. Essas
representaes vetoriais, que correspondem aos pesos da camada oculta
dessa rede neural, so obtidas a partir da otimizao de uma funo de
custo. O diferencial dessa tcnica  que os vetores aprendidos j so
teis, mesmo sem se construir o modelo de rede neural por completo.

O importante a se saber sobre essa tcnica  que as representaes
distribudas[4] obtidas funcionam com qualquer representao de
linguagem cuja formao possa ser formulada como uma tarefa de
otimizao de modelo de probabilidade. As representaes distribudas

ocorrncia de cada palavra no texto a partir da probabilidade de suas

As representaes de palavras calculadas usando a tcnica acima

No entanto, essa tcnica no permite que se gere representaes de

Representao de distribuies de textos completos
--------------------------------------------------

Pesquisadores comearam ento a procurar maneiras de adaptar as
representaes distribudas de palavras para criar representaes de
textos completos, ou grandes volumes de textos, dada sua utilidade para,
por exemplo, classificao de documentos. A forma proposta por (LE;
MIKOLOV, 2014), e usada neste trabalho, consiste em se adicionar rtulos
aos documentos e deix-los influenciar a distribuio de probabilidade
das demais palavras, como se cada rtulo estivesse prximo de todas as
palavras dos documentos em que figuram. O produto  uma representao
vetorial que localiza cada documento dentro do mesmo espao vetorial
compartilhado com as representaes de suas palavras. Essas
representaes podem ento ser usadas nas mesmas operaes algbricas
ilustradas em **Representao de distribuies de palavras**, mas tambm
como entrada para vrios outros algoritmos e tcnicas, como se ver
adiante.

Partiu-se da premissa que PPGs que investigam e produzem pesquisa
(representadas por suas teses e dissertaes) nos mesmos campos ou em
campos muito similares, compartilham um mesmo vocabulrio prprio, ou
seja, termos e palavras cujos conceitos caracterizam o domnio de
conhecimento a que pertencem, de uma forma particular. H termos que
podem ser considerados muito caractersticos da Astronomia porque
dificilmente o encontraremos em textos cientficos de qualquer outra
rea de conhecimento, e vice-versa. Esses termos estaro muito prximos
no espao semntico de outros junto aos quais frequentemente aparecem
nos textos. Por outro lado, h termos frequentes que aparecem em todas
as reas de conhecimento. Esses termos acabam por receber representaes
que os colocam ao mesmo tempo equidistantes, e por isso distantes dos
temas mais especializados. Significa que termos altamente
especializados, especficos de uma rea de conhecimento, so capazes de
proporcionar o rpido reconhecimento da rea a que pertence um texto
pois ocupam uma mesma regio no espao semntico criado por tais
modelos.

A tcnica de (LE; MIKOLOV, 2014) usada neste trabalho, , em resumo, uma
tcnica de representao distribuda no supervisionada (no assistida
por humanos e classificaes previas por eles providas) que aprende
representaes vetoriais de textos dentro do mesmo espao ocupado pelas
representaes das suas palavras. Lembrando que essa representao de
textos no  resultado de uma simples operao (ex.: mdia ou soma) das
representaes de suas palavras componentes, mas uma distribuio de
probabilidade em si.

Multidimensional Scaling (*MDS*)
--------------------------------

O escalonamento multidimensional (MDS) refere-se a um conjunto de
tcnicas de ordenao usadas para a visualizao de informaes, em
particular, para exibir informaes contidas em uma matriz de
distncias. , em sntese, uma forma de reduo de dimensionalidade no
linear.  tambm um meio de visualizar o nvel de similaridade de casos
individuais dentro de um conjunto de dados.

Em termos mais tcnicos, dada uma matriz com as distncias entre cada
par de objetos em um conjunto, e um nmero escolhido de dimenses N, um
algoritmo MDS coloca cada objeto no espao N-dimensional de forma que as
distncias entre os objetos sejam preservadas da melhor forma possvel.
Por essa razo, o MDS  usado para traduzir "informaes sobre as
'distncias' de pares entre um conjunto de n objetos ou indivduos" em
uma configurao de n pontos mapeados em um espao cartesiano abstrato.
Esses espaos de destino podem ter qualquer dimensionalidade. No
entanto, usualmente se opta por 2 ou 3 dimenses para ser mais
facilmente assimilvel e compatvel com programas de gerao de imagens.
